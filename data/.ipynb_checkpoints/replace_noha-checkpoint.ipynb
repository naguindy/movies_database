```python
import pandas as pd
```

```python
data = pd.read_csv('tags_adj.csv')
```

```python
data.head()
```

```python
# Using 'assign' to create a new column with altered values
```

```python
df2 = data.assign(result=data['adjectives'].str.replace(r'\[', ''))
```

```python
df2.head()
```

```python
df3 = df2.assign(result2=df2['result'].str.replace(r'\]', ''))
```

```python
df3
```

```python
df4 = df3.assign(result3=df3['result2'].str.replace(r'\'', ''))
```

```python
df4
```

```python
tag_adj = df4.drop(['adjectives', 'result', 'result2', ''], axis =1)
```
